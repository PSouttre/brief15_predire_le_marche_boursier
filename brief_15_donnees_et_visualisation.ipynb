### Comparaison des différentes sociétés

```python
plt.figure(figsize=(12, 6))

for entreprise in ['APPLE', 'MICROSOFT', 'AMAZON', 'ALPHABET', 'NVIDIA', 'META', 'TESLA']:
    df = data[entreprise].copy()
    df['date'] = pd.to_datetime(df.index)
    plt.plot(df['date'], df['close'], label=entreprise)

plt.title("Comparaison des prix de clôture")
plt.xlabel("Date")
plt.ylabel("Cours (USD)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(16, 10))
for i, name in enumerate(data.keys(), 1):  # Les 7 entreprises
    plt.subplot(3, 3, i)  # 3 lignes, 3 colonnes 
    data[name]["close"].plot()
    plt.title(name)
    plt.ylabel("close")
    plt.xlabel(None)

plt.tight_layout()
plt.show()
```

### Recherche de tendance et de saisonnalité sur les données Apple

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt

company_name = "APPLE"  r

# Série temporelle 
ts = data[company_name]["close"]

# Décomposition multiplicative de la série
decomposition = sm.tsa.seasonal_decompose(ts, model='multiplicative', period=252)  # 'period=252' pour une saisonnalité annuelle avec des données journalières

# Visualisation des composants
plt.figure(figsize=(12, 8))
decomposition.plot()
plt.show()
```

Tendance à la hausse

Présence d'une saisonnalité

```python
from statsmodels.tsa.stattools import adfuller

# Test de Dickey-Fuller pour la stationnarité
result = adfuller(ts)
print(f"p-value: {result[1]}")
```

La série n'est pas stationnaire : il est préférable d'appliquer une différenciation avant d'utiliser un modèle comme ARIMA.

```python
# Calcul de la différence entre chaque valeur et la valeur précédente de la série + suppression des valeurs NaN crées
ts_diff = ts.diff().dropna()
```

```python
result = adfuller(ts_diff)
f"p-value après différenciation : {result[1]}"
```

La p-value est encore importante, la série n'est pas encore stationaire

```python
# Différenciation saisonnière
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

decomposition = seasonal_decompose(ts_diff, model='additive', period=1)  # périodicité annuelle
# Affichage des composantes
plt.figure(figsize=(14, 8))
decomposition.plot()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))

# Série originale
plt.subplot(1, 2, 1)
plt.plot(ts, label='Série originale')
plt.title("Série temporelle originale")
plt.xlabel("Date")
plt.ylabel("Valeur")
plt.grid(True)
plt.legend()

# Série différenciée
plt.subplot(1, 2, 2)
plt.plot(ts_diff, label='Série différenciée', color='orange')
plt.title("Série après différenciation")
plt.xlabel("Date")
plt.ylabel("Différence")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
